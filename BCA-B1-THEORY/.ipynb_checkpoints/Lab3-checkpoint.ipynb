## 1) Numeric Data Types in Python

Python provides three main numeric data types:

- `int`     → Whole numbers (no decimal)
- `float`   → Decimal numbers (real numbers)
- `complex` → Complex numbers like `a + bj`

Additionally, `bool` is also treated as a numeric type in Python.

```python

a = 10
b = -99
c = 0

print("a =", a, "| type:", type(a))
print("b =", b, "| type:", type(b))
print("c =", c, "| type:", type(c))

print("a + b =", a + b)
print("a * 5 =", a * 5)
```

***
### `float` (Floating Point)

Floats are decimal numbers.  
Python floats are stored in binary floating format, so small precision issues may occur.
***


```python

x = 10.5
y = -3.14

print("x =", x, "| type:", type(x))
print("y =", y, "| type:", type(y))
print("x + y =", x + y)
print("x * 2 =", x * 2)
```

```python

print("0.1 + 0.2 =", 0.1 + 0.2)
print("Is 0.1 + 0.2 equal to 0.3?", (0.1 + 0.2) == 0.3)
```

***
### `complex`: Complex Numbers

Complex numbers are represented as:

`a + bj`  
where:
- `a` = real part
- `b` = imaginary part  
- `j` = √(-1) in Python
***


```python

z = 2 + 3j

print("z =", z, "| type:", type(z))
print("Real part:", z.real)
print("Imaginary part:", z.imag)
```

```python

z1 = 1 + 2j
z2 = 3 + 4j

print("z1 =", z1)
print("z2 =", z2)
print("z1 + z2 =", z1 + z2)
print("z1 * z2 =", z1 * z2)
```

***
### `bool`(Boolean) as Numeric Type

In Python:
- `True` behaves like `1`
- `False` behaves like `0`
***

```python

print("True + 5 =", True + 5)
print("False * 10 =", False * 10)
print("type(True) =", type(True))
print("int(True) =", int(True))
print("int(False) =", int(False))
```

***
### Type Casting (Conversions)
Python allows conversion between numeric types:

- `int("10")` → 10  
- `float("10")` → 10.0  
- `complex(2, 5)` → (2+5j)  
- `bool(0)` → False  
- `bool(100)` → True  
***

```python
s = "10"

i = int(s)
f = float(s)
comp = complex(2, 5)

print("s =", s, "| type:", type(s))
print("int(s) =", i, "| type:", type(i))
print("float(s) =", f, "| type:", type(f))
print("complex(2, 5) =", comp, "| type:", type(comp))

print("bool(0) =", bool(0))
print("bool(100) =", bool(100))
```

***
### Useful Numeric Built-in Functions

- `abs()`   → absolute value  
- `pow()`   → power  
- `round()` → rounding  
- `max()`   → maximum value  
- `min()`   → minimum value  
 ***

```python
print("abs(-25) =", abs(-25))
print("pow(2, 5) =", pow(2, 5))
print("round(3.14159, 2) =", round(3.14159, 2))
print("max(10, 20, 5) =", max(10, 20, 5))
print("min(10, 20, 5) =", min(10, 20, 5))
```

***
### Arithmetic Operators

- `+` Addition  
- `-` Subtraction  
- `*` Multiplication  
- `/` Division (always float)  
- `//` Floor division  
- `%` Modulus (remainder)  
- `**` Power
***

```python

a = 10
b = 3

print("a =", a, "b =", b)
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

***Ques 2***: Write a Python program to check whether a given number is an Armstrong number or not.

```python
num = int(input("Enter a number: "))
n = num
power = len(str(num))
total = 0

while n > 0:
    digit = n % 10
    total += digit ** power
    n //= 10

if total == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

***Explanation:***

-	The program first takes a number as input and stores it in num.
-   The total number of digits in the number is calculated using len(str(num)) and stored in power.
-	A copy of the original number is stored in n so that the original value is not lost.
-	The variable total is initialized to 0 to store the sum of powered digits.
-	Using a while loop, each digit of the number is extracted from right to left using the modulus operator % 10.
-	Each extracted digit is raised to the power of the total number of digits and added to total.
-	The number is reduced by one digit in each iteration using integer division // 10.
-	After the loop ends, the value of total is compared with the original number.
-	If both values are equal, the number is an Armstrong Number; otherwise, it is not an Armstrong Number.

***Ques 3***: Write a Python program to check whether a given number is a palindrome or not.

```python
n=int(input("Enter number:"))
temp=n
rev=0
while(n>0):
    dig=n%10
    rev=rev*10+dig
    n=n//10
if(temp==rev):
    print("The number is a palindrome!")
else:
    print("The number isn't a palindrome!")
```

***Explanation:*** 
-    The program first takes a number as input and stores it in n.
-    A copy of the original number is saved in temp so it can be compared later.
-    The variable rev is initialized to 0 to store the reversed number
-    Using a while loop, the program extracts the last digit of the number using the modulus operator % 10.
-    The extracted digit is added to rev after shifting its existing digits one place to the left using rev = rev * 10 + dig.
-    The original number is reduced by one digit in each iteration using integer division // 10.
-    The loop continues until all digits of the number are processed.
-    After the loop ends, the reversed number stored in rev is compared with the original number stored in temp.
-    If both values are equal, the number is a Palindrome Number; otherwise, it is not a Palindrome Number.
