# Introduction to Python

#### Python was created by Guido van Rossum and released in 1991. It focuses on readability and simplicity, which means you can write programs using fewer lines of code compared to many other languages. Python is open-source and runs on Windows, macOS, and Linux

## **Questions** 

Q1. What is interactive programming in Python?

Ans- Interactive programming means writing and executing Python code line by line and getting immediate output. It is mainly used for learning, testing small pieces of code, and debugging. The user interacts directly with the Python interpreter.

Advantages:
- Immediate results
- Easy to test logic
- Beginner-friendly

Q2. What is script programming in Python?


Ans- Script programming involves writing Python code in a file (usually with a .py extension) and executing the entire file at once. It is mainly used for developing complete programs and applications.

Advantages:
- Suitable for large programs
- Code can be reused
- Easier to maintain

> ## Dynamic Types in Python

Python uses dynamic typing, which means you do not need to declare the type of a variable explicitly. The type is decided automatically based on the value assigned to the variable.


```python
print("x = 10 → integer")
print("x = 'Hello' → string")
```

Q. What is the command to check the data type?

The built-in function type() is used to check the data type of a variable.

Q. What are immutable data types?

Immutable data types are those whose values cannot be changed after creation. If you try to modify them, a new object is created instead.

**Examples:**

- int
- float
- string
- tuple

Q. What is a tuple?

A tuple is an ordered collection of elements enclosed in parentheses (). Tuples are immutable, meaning their values cannot be changed once created.

```python
print("my_tuple = (1, 2, 3)")
```

## Variable – Native Python
*** 

Variables in Python are used to store data values. A variable name acts as a reference to the data stored in memory.

## Variable Definition

A variable is a named location used to store data in memory. The value of a variable can be changed during program execution.

> ## Naming and Using Variables

**Rules for naming variables:**

- Must start with a letter or underscore
- Cannot start with a number
- No spaces allowed
- Should not be a Python keyword


**Note on Keywords**

Python has reserved words called keywords which have special meanings. These should not be used as variable names to avoid errors and confusion.

## Variable Types
---

- int (numbers)
- float (decimal numbers)
- str (text)
- bool (True or False)

# String

A string is a sequence of characters enclosed within single quotes or double quotes. Strings are used to store text data.

```python
print("name = python")
```

# Python String Built-in Methods

Python provides many built-in methods to work with strings. These methods help in modifying, searching, and formatting string data.

> ## Case Conversion Methods

- upper() converts all characters in a string to uppercase.
- lower() converts all characters in a string to lowercase.
- title() converts the first character of each word to uppercase.
- capitalize() converts only the first character of the string to uppercase.
- swapcase() swaps uppercase letters to lowercase and vice versa.    

> ## Remove Spaces / Cleanup

- strip() removes spaces from both the beginning and end of a string.
- lstrip() removes spaces from the left side of a string.
- rstrip() removes spaces from the right side of a string.
