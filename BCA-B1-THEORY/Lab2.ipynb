### <span style="color: rgb(0, 128, 0);">Python is an **Object-Oriented Language**</span>

In Python, **everything is treated as an object**.

This includes:

- integers → `10`  
- strings → `"hello"`  
- lists → `[1, 2, 3]`  
- functions → `print`, `len`

- even modules → `math`, `os`

***

**Hence, in Python:**

> Data and behavior are bundled together using objects.

### What is an Object in Python?
**(Theoretical Definition)**

An object is a runtime entity that has:

- **Identity** → a unique memory address (`id()`)  
- **Type** → specifies the kind of object (`type()`)  
- **State / Value** → the actual data stored inside the object

### Example

````python
x = 10  
print(id(x)) # identity  
print(type(x)) # type  
print(x) # value
````

### Internally in Python
- `10` is an object created from the class `int`
- The class `int` itself is also an object (a class object)

# <span style="color: rgb(0, 128, 0);">Built-in Functions vs Methods (Internal Working)</span>

## A) Built-in Function (Global / independent)

Example:

````python
len("python")
````

###  Internal meaning:

Python searches `len` inside the **built-in namespace**.

Built-ins live in:

- `__builtins__`
- `builtins` module

So this is basically:

```python
builtins.len("python")
```

Built-in functions are **not tied to one object**.

* * *

## Method (belongs to an object’s class)

Example:

````python
"python".upper()
````

###  Internal meaning:

Python does:

1.  look at object type → `str`
2.  find `upper` inside `str`
3.  bind it with the object `"python"`
4.  then call it

Equivalent internal call:

```python
str.upper("python")
```

So, **method = function that becomes bound to an object**.

* * *

# <span style="color: rgb(186, 55, 42);">Why Methods Feel Different?</span>

Because in methods:

### Python automatically passes the object as the first argument

That object is called:

- `self` (conventionally)

Example:

````python
class Demo:
    def show(self):
        print("Hello")

d = Demo()
d.show()
````

Internally:

```python
Demo.show(d)
```

So:

> `d.show()` is just a shortcut for `Demo.show(d)`

Because of **design**:

### `len()` is general-purpose

- list

```python
obj.__len__()

Internally becomes:

```python

`len()` is a built-in function that **calls the method indirectly**.

# Variable – Native Python

A variable in Python is a named location in memory used to store data.  
The value stored in a variable can change while the program is running.

### Naming and Using Variables — Guidelines

- Variable names may contain letters, numbers, and underscores.
  They must start with a letter or an underscore, not a number.
  Example: message_1 is valid, 1_message is invalid.

- Spaces are not allowed in variable names.
  Use underscores to separate words.
  Example: greeting_message works, greeting message causes an error.

- Do not use Python keywords or built-in function names as variable names.
  Words such as print, list, or type have special meanings in Python.

- Variable names should be short but descriptive.
  Example: name is better than n,
  student_name is better than s_n.

- Avoid using lowercase letter l and uppercase letter O
  because they can be confused with the numbers 1 and 0.

### NOTE

At this stage, it is recommended to use lowercase letters for variable names.
Python allows uppercase letters, but avoiding them now helps in writing clear and readable code.

`casefold()` (stronger lower)

### Definition and Usage

Return a casefolded copy of the string. Casefolded strings may be used for caseless matching.

Casefolding is similar to lowercasing but more aggressive because it is intended to remove all case distinctions in a string. For example, the German lowercase letter `'ß'` is equivalent to `"ss"`. Since it is already lowercase, [`lower()`](https://docs.python.org/3/library/stdtypes.html#str.lower "str.lower") would do nothing to `'ß'`; [`casefold()`](https://docs.python.org/3/library/stdtypes.html#str.casefold "str.casefold") converts it to `"ss"`. For example:

````python
"GERMAN ß".casefold() 
````

## <span style="color: rgba(11, 46, 69, 1);">2) Remove Spaces / Cleanup</span>

`strip(chars=None)`

### Definition and Usage

Return a copy of the string with the leading and trailing characters removed. The *chars* argument is a string specifying the set of characters to be removed. If omitted or `None`, the *chars* argument defaults to removing whitespace. The *chars* argument is not a prefix or suffix; rather, all combinations of its values are stripped:


* * *

````python
"  hello      vaanya    ".strip()   # 'hello'
````

***
`lstrip(chars=None)`

### Definition and Usage

Return a copy of the string with leading characters removed. The *chars* argument is a string specifying the set of characters to be removed. If omitted or `None`, the *chars* argument defaults to removing whitespace. The *chars* argument is not a prefix; rather, all combinations of its values are stripped:

* * *

````python
"  hello".lstrip()    # 'hello'
````

* * *

`rstrip(chars=None)`

### Definition and Usage

Return a copy of the string with trailing characters removed. The *chars* argument is a string specifying the set of characters to be removed. If omitted or `None`, the *chars* argument defaults to removing whitespace. The *chars* argument is not a suffix; rather, all combinations of its values are stripped:

* * *

````python
"hello   ".rstrip()   # 'hello'
````

## find(sub[, start[, end]])

### Definition and Usage

The `find()` method searches for a substring inside a string and returns the
index of the first occurrence of the substring.

The search is performed from left to right.
If the substring is found, the starting index is returned.
If the substring is not found, the method returns `-1`.

The optional `start` and `end` parameters are used to limit the search
to a specific portion of the string.

***

### Syntax

```python
string.find(sub, start, end)

````python
"python".find("th")
````

````python
"python".find("zz")
````

````python
"banana".find("a", 2)
````

***
## rfind(sub[, start[, end]])
***

### Definition and Usage

The `rfind()` method searches for a substring in a string starting from
the **right side** and moving towards the left.

Although the search begins from the right, the index returned is still
based on the normal left-to-right indexing system.
If the substring is found, the method returns the **highest index**
at which the substring starts.
If the substring is not found, it returns `-1`.

This method is useful when you want to find the **last occurrence**
of a substring without causing an error.

***

### Syntax

```python
string.rfind(sub, start, end)

````python
"banana".rfind("a")
````

````python
"banana".rfind("x")
````

***
## index(sub[, start[, end]])
***

### Definition and Usage

The `index()` method searches for a substring in a string and returns
the index of the **first occurrence**, similar to the `find()` method.

The key difference is in error handling.
If the substring is not found, the `index()` method raises a
`ValueError` instead of returning `-1`.

This method is useful when the substring is **expected to be present**
and its absence should cause the program to stop immediately.

***

### Syntax

```python
string.index(sub, start, end)

````python
"python".index("th")
````

````python
"python".index("zz")
````

***
## rindex(sub[, start[, end]])
***

### Definition and Usage

The `rindex()` method is the reverse version of the `index()` method.

It searches for a substring starting from the **right side** of the string
and returns the index of the **last occurrence**.
If the substring is not found, the method raises a `ValueError`.

This method is useful when you need the last match
and want strict validation, meaning the substring must exist.

***

### Syntax

```python
string.rindex(sub, start, end)

````python
"banana".rindex("a")
````

***
## count(sub[, start[, end]])
***

### Definition and Usage

The `count()` method is used to count the number of times
a substring appears in a string.

It counts only **non-overlapping occurrences**
and returns the total number of matches.
This method does not modify the original string.

The `count()` method is useful when you need frequency information,
such as how many times a word, character, or pattern
exists within a given text.

The optional `start` and `end` parameters allow counting
to be limited to a specific portion of the string.

***

### Syntax

```python
string.count(sub, start, end)

````python
"banana".count("a")
````

````python
"banana".count("an")
````

***
## startswith(prefix[, start[, end]])
***

### Definition and Usage

The `startswith()` method checks whether a string begins
with a specified prefix.

It returns `True` if the prefix matches the beginning of the string.
Otherwise, it returns `False`.
This method checks only the starting position of the string
and does not search the entire string.

The `startswith()` method is commonly used for validation tasks,
such as checking file names, URL prefixes like `http` or `https`,
and enforcing naming conventions.

***

### Syntax

```python
string.startswith(prefix, start, end)

````python
"python".startswith("py")
````

````python
"python".startswith("on")
````

***
## endswith(suffix[, start[, end]])
***

### Definition and Usage

The `endswith()` method checks whether a string ends
with a specified suffix.

It returns `True` if the suffix matches the ending of the string.
Otherwise, it returns `False`.
This method checks only the ending position of the string
and does not scan the entire string.

The `endswith()` method is commonly used for file validation,
such as checking file extensions like `.txt`, `.csv`, or `.pdf`,
and for verifying fixed naming patterns.

***

### Syntax

```python
string.endswith(suffix, start, end)

````python
"python".endswith("on")
````

````python
"python".endswith("py")
````
