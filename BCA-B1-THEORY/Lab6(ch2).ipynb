### 1) Initialization

* Strings are **immutable** sequences of characters.
* Create using:

  * Single quotes: `'hello'`
  * Double quotes: `"hello"`
  * Triple quotes (multi-line): `"""hello"""`

```python
s1 = 'Python'
s2 = "Data Science"
s3 = """Line1
Line2"""

print(s1)
print(s2)
print(s3)
```

### 2) Operators on strings

* `+` concatenation, `*` repetition
* `in` membership
* comparison uses lexicographic order

```python
print("Py" + "thon")      # 'Python'
print("ha" * 3)           # 'hahaha'
print("a" in "cat")       # True
print("abc" < "abd")      # True
```

### 3) Indexing (0-based)

```python
s = "Python"
print(s[0])   
print(s[-1])  # 
print(s[len(s)-1])
print(s[len(s)])
```

### 4) Slicing

Form: `s[start : stop : step]` (stop excluded)

```python
s = "Python"
print(s[1:4])   
print(s[:3])     
print(s[3:])     
print(s[::-1])   
```

### 5) `split()`

Splits into list by delimiter (default = whitespace)


```python
print("one two three".split())         
print("2026-02-10".split("-"))
```

> # Lists

### 1) Initialization

Lists are **mutable**, ordered collections.

```python
L = [10, 20, 30]
mixed = [1, "A", 3.5, True]
empty = []
print(L)
print(mixed)
print(empty)
```

### 2) Common methods

* `append(x)`, `extend(iterable)`, `insert(i,x)`
* `remove(x)`, `pop(i)`, `clear()`
* `index(x)`, `count(x)`
* `sort()`, `reverse()`

```python
L = [3, 1, 2]
L.append(5)         
print(L)
L.sort()            
print(L)
L.pop()
print(L)
```

### 3) Operations

* Concatenation, repetition, membership
* indexing/slicing same as strings

```python
print([1,2] + [3,4])      
print([0]*4)              
print(2 in [1,2,3]) 
```

### 4) List comprehension

Compact way to build lists.

```python
squares = [x*x for x in range(1,6)]
evens = [x for x in range(10) if x%2==0]
pairs = [(x,y) for x in [1,2] for y in [3,4]]
print(squares)
print(evens)
print(pairs)
```

### 5) Nesting

List of lists (2D structure).

```python
mat = [[1,2,3],[4,5,6]]
print(mat[1][2])  
```

> # Tuples

### 1) Initialization

Tuples are **immutable** ordered collections.

```python
t1 = (1,2,3)
t2 = 1,2,3          
t3 = (5,)           
print(t1)
print(t2)
print(t3)
```

### 2) Operations

* indexing/slicing, `+`, `*`, `in`

```python
t = (10,20,30,40)
print(t[1])      
print(t[1:3])
```

### 3) Methods (limited)

* `count(x)`, `index(x)`

```python
print((1,2,2,3).count(2))  
print((1,2,2,3).index(2)) 
```

### 4) Nesting

```python
T = ((1,2),(3,4))
print(T[1][0])  
```

### 5) List vs Tuple

| Feature    | List         | Tuple                |
| ---------- | ------------ | -------------------- |
| Mutability | Mutable      | Immutable            |
| Speed      | Slower       | Faster               |
| Methods    | Many         | Few                  |
| Use case   | dynamic data | fixed records / keys |

> # Sets & Dictionaries

## A) Sets

### 1) Initialization

Sets are **unordered**, **unique** elements.

```python
S = {1,2,3}
S2 = set([1,2,2,3])   
empty_set = set()     
print(S)
print(S2)
print(empty_set)
```

### 2) Methods / Operations

* `add`, `remove`, `discard`, `pop`
* `union(|)`, `intersection(&)`, `difference(-)`, `symmetric_difference(^)`

```python
A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
print(A | B)   
print(A & B)   
print(A - B) 
```

## B) Dictionaries

### 1) Initialization

Key–value mapping. Keys must be hashable (immutable types).

```python
d = {"name":"Vaanya", "age":18}
d2 = dict(a=1, b=2)
print(d)
print(d2)
```

### 2) Common operations/methods

* Access: `d[key]`, safe access: `d.get(key, default)`
* Insert/update: `d[key] = value`, `update()`
* Remove: `pop(key)`, `popitem()`, `del d[key]`
* Views: `keys()`, `values()`, `items()`

```python
print(d)
d["age"] = 39
print(d)
print(d.get("city", "NA"))
```

### 3) Nesting

```python
student = {
  "name":"A",
  "marks":{"math":90, "python":95}
}
student["marks"]["python"]   
```

### 4) Sorting & typecasting

* Sort dict by keys or by values using `sorted()`

```python
d = {"b":2, "a":5, "c":1}
print(sorted(d))                     
print(sorted(d.items()))             
print(sorted(d.items(), key=lambda x: x[1]))
```

### 4) Scope
----

* #### Local, Enclosing, Global, Built-in (LEGB)

#### Python follows the LEGB Rule for Scope
Python looks for a variable in this order:

- > ### L – Local Scope
    •	Variables created inside a function
	•	Can be used only inside that function

- > ### E – Enclosing Scope
	•	Variables in a function inside another function
	•	Used in nested functions
- > ### G – Global Scope
	•	Variables created outside all functions
	•	Can be used anywhere in the program



```python
x = 10
def foo():
    x = 5
    return x
```

```python
x=10
print(foo())
print(x)
```

### 5) Recursion

Function calls itself; must have base case.

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
   
```

```python
print(fact(4))
```

> # Advanced Functions


### 1) `lambda`

Anonymous function.

```python
square = lambda x: x*x
print(square(2))
```

### 2) `map()`

Apply a function to every element.

```python
nums = [1,2,3]
list(map(lambda x: x*x, nums))  
```

### 3) `filter()`

Keep only elements satisfying condition.

```python
nums = [1,2,3,4,5]
list(filter(lambda x: x%2==0, nums)) 
```
