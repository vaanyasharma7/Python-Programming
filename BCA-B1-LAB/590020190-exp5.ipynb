> ## Strings and Sets

### Q1) Count number of capital letters in a string


```python
text = input("Enter a string: ")
count = 0

for ch in text:
    if ch.isupper():
        count += 1

print("Number of capital letters:", count)
```

### Q2) Count total number of vowels in a string

```python
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print("Number of vowels:", count)
```

### Q3) Input a sentence and print words in separate lines

```python
sentence = input("Enter a sentence: ")
words = sentence.split()

for word in words:
    print(word)
```

### Q4) Count occurrences of a substring in a string

```python
string = input("Enter the string: ")
sub = input("Enter substring: ")
count = 0

for i in range(len(string) - len(sub) + 1):
    if string[i:i+len(sub)] == sub:
        count += 1

print(count)
```

### Q5) Count occurrences of each alphabet (case insensitive)


```python
text = input("Enter a string: ")
text = text.upper()

for ch in sorted(set(text)):
    if ch.isalpha():
        print(text.count(ch), ch)
```

### Q6) Count number of unique words using sets

```python
sentence = input("Enter a sentence: ")
words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
```

### Q7) Set operations on fruits

```python
s1 = set(input("Enter fruits for set 1: ").split())
s2 = set(input("Enter fruits for set 2: ").split())

print("Common fruits:", s1 & s2)
print("Only in s1:", s1 - s2)
print("Total fruits count:", len(s1 | s2))
```

### Q8) Perform set operations



```python
S1 = {"red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference:", S1 - S2)
```
