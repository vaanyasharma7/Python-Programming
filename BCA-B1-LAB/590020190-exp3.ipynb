## Conditional Statements

#### 1. Check whether a given number is divisible by both 3 and 5

```python
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Number is divisible by both 3 and 5")
else:
    print("Number is NOT divisible by both 3 and 5")
```

*** 
- Takes an integer input from the user
- Uses modulo operator % to check divisibility
- Applies logical and condition
- Prints result based on condition
***

#### Q3. Find the greatest among two numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest number:", a)
elif b > a:
    print("Greatest number:", b)
else:
    print("Numbers are equal")
```

#### 5. Check whether quadratic equation has real or imaginary roots and display them

```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Roots are real and different")
    print("Root 1:", r1)
    print("Root 2:", r2)

elif d == 0:
    r = -b / (2*a)
    print("Roots are real and same")
    print("Root:", r)

else:
    real = -b / (2*a)
    imag = math.sqrt(-d) / (2*a)
    print("Roots are imaginary")
    print("Root 1:", real, "+", imag, "i")
    print("Root 2:", real, "-", imag, "i")
```

***
- Takes coefficients a, b, and c as input
- Calculates discriminant using formula bÂ² - 4ac
- Checks discriminant value using conditional statements
- Displays real or imaginary roots accordingly
***

#### 6. Check whether a given year is a leap year

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

*** 
- Accepts year as input
- Uses leap year rules with % operator
- Applies logical or and and
- Prints whether year is leap year or not
***

#### 7. Display the next date of the calendar

```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# check leap year
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days_in_month[1] = 29

day += 1

if day > days_in_month[month - 1]:
    day = 1
    month += 1

    if month > 12:
        month = 1
        year += 1

print("Next Date:")
print("Day:", day, "Month:", month, "Year:", year)
```

***
- Takes day, month, and year as input
- Checks number of days in a month
- Adjusts date for month and year change
- Prints the next valid date

#### 8. Print grade sheet of a student using CGPA


```python
name = input("Enter student name: ")
roll = input("Enter roll number: ")

m1 = int(input("Enter marks 1: "))
m2 = int(input("Enter marks 2: "))
m3 = int(input("Enter marks 3: "))
m4 = int(input("Enter marks 4: "))
m5 = int(input("Enter marks 5: "))

total = m1 + m2 + m3 + m4 + m5
percentage = total / 5
cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"

print("\n--- Grade Sheet ---")
print("Name:", name)
print("Roll No:", roll)
print("Total Marks:", total)
print("Percentage:", percentage)
print("CGPA:", cgpa)
print("Grade:", grade)
```

***
- Takes day, month, and year as input
- Checks number of days in a month
- Adjusts date for month and year change
- Prints the next valid date
