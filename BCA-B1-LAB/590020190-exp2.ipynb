2. Find area of a circle (radius as input)

### ***Program to find the area of a circle***
- Radius is taken as input
- Formula: Area = π × r²

```python
radius = float(input("Enter radius: "))

area = 3.14 * radius * radius
print("Area of Circle:", area)
```

6. Program to find hypotenuse using Pythagoras theorem

### ***Formula: c = √(a² + b²)***

```python
import math

a = float(input("Enter base: "))
b = float(input("Enter height: "))

c = math.sqrt(a*a + b*b)
print("Hypotenuse:", c)
```

7. Program to calculate Simple Interest

### ***Formula: SI = (P × R × T) / 100***

```python
p = float(input("Enter principal amount: "))
r = float(input("Enter rate of interest: "))
t = float(input("Enter time: "))

si = (p * r * t) / 100
print("Simple Interest:", si)
```

8. Program to find area of triangle (sides given)

### ***Using Heron’s Formula***

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Area of Triangle:", area)
```

9. Program to convert seconds into hours, minutes and seconds

```python
seconds = int(input("Enter total seconds: "))

hours = seconds // 3600
seconds = seconds % 3600
minutes = seconds // 60
seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", seconds)
```

10. Program to swap two numbers without using third variable

### ***Swapping using tuple assignment***

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a, b = b, a

print("After Swapping")
print("a =", a)
print("b =", b)
```

11. Program to find sum of first n natural numbers

### ***Using formula n(n+1)/2***

```python
n = int(input("Enter value of n: "))

sum_n = n * (n + 1) // 2
print("Sum of first", n, "natural numbers:", sum_n)
```

12. Program to print truth table for bitwise operators

### ***Operators used: &, |, ^***

```python
a = 1
b = 0

print("a =", a, "b =", b)
print("a & b =", a & b)
print("a | b =", a | b)
print("a ^ b =", a ^ b)
```

13. Program to find left shift and right shift values

```python
num = int(input("Enter a number: "))

print("Left Shift:", num << 1)
print("Right Shift:", num >> 1)
```

14. Program to check membership of a number in a sequence

### ***Sequence(given): (10, 20, 56, 78, 89)***

```python
sequence = (10, 20, 56, 78, 89)
num = int(input("Enter number: "))

if num in sequence:
    print("Number is present in sequence")
else:
    print("Number is not present in sequence")
```

15. Program to check membership of a character in a string
