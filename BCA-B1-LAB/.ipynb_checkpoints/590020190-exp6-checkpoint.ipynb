> ## Lists, Tuples and Dictionary

### Q1) Scan values in range 0â€“3 and count occurrences

```python
n = int(input("Enter number of elements: "))
nums = []

for i in range(n):
    nums.append(int(input()))

for i in range(4):
    print(i, "occurred", nums.count(i), "times")
```

### Q2) Create a tuple and find average

```python
n = int(input("Enter number of values: "))
values = []

for i in range(n):
    values.append(int(input()))

t = tuple(values)
print("Average:", sum(t) / len(t))
```

### Q3) Find runner-up score

```python
scores = list(map(int, input("Enter scores: ").split()))
unique_scores = list(set(scores))
unique_scores.sort(reverse=True)

print("Runner-up score:", unique_scores[1])
```

### Q4) Dictionary of students and cities

```python
n = int(input("Enter number of students: "))
students = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    students[name] = city

print("Names:", students.keys())
print("Cities:", students.values())

for name, city in students.items():
    print(name, "->", city)
```

### Q5) Movie details dictionary

```python
n = int(input("Enter number of movies: "))
movies = {}

for i in range(n):
    name = input("Movie name: ")
    year = int(input("Year: "))
    director = input("Director: ")
    cost = int(input("Production cost: "))
    earning = int(input("Collection: "))

    movies[name] = {
        "year": year,
        "director": director,
        "cost": cost,
        "earning": earning
    }

for m, d in movies.items():
    print(m, d)

print("Movies before 2015:")
for m, d in movies.items():
    if d["year"] < 2015:
        print(m)

print("Profitable movies:")
for m, d in movies.items():
    if d["earning"] > d["cost"]:
        print(m)
```
