### Q1) Find the factorial of a given number

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial is:", fact)
```

### Q2) Check whether the given number is an Armstrong number

```python
num = int(input("Enter a number: "))
temp = num
power = len(str(num))
sum_val = 0

while temp > 0:
    digit = temp % 10
    sum_val += digit ** power
    temp //= 10

if sum_val == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

### Q3) Print Fibonacci series up to given term



```python
n = int(input("Enter number of terms: "))
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

### Q4) Check whether the given number is prime or not

```python
num = int(input("Enter a number: "))
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break

if flag:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

### Q5) Check whether the given number is palindrome or not

```python
num = input("Enter a number: ")

if num == num[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

### Q6) Print sum of digits of a number

```python
num = int(input("Enter a number: "))
sum_digits = 0

while num > 0:
    sum_digits += num % 10
    num //= 10

print("Sum of digits:", sum_digits)
```

### Q7) Count and print all numbers divisible by 5 or 7 between 1 and 100 

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i)
        count += 1

print("Count:", count)
```

### Q8) Convert all lowercase letters to uppercase in a string

### Q9) Print all prime numbers between 1 and 100

```python
for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num)
```

### Q10) Print multiplication table of a given number

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "*", i, "=", num * i)
```
