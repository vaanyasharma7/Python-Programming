## Python Installation and starting with python

### 1. Install Python & Difference Between Scripting and Interactive Mode (IDLE)


Python Installation

- Download Python from the official website.
- During installation, tick “Add Python to PATH”.
- IDLE is installed automatically with Python.

Interactive Mode

- Executes code line by line
- Used for quick testing
- Prompt: >>>

Scripting Mode

- Used to write full programs
- Code saved with .py extension
- Program runs at once
