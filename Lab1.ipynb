```python
GREEN = "\033[92m"
RESET = "\033[0m"
```

# Introduction

```python
print("Python is a high-level, easy-to-learn programming language.")
print("It can be used for scripting, web development, and data analysis.")
print("Python code is simple and readable, which makes it great for beginners.")
print("It supports variables, loops, functions, and more.\n")
```

# Question 1

```python
print("Q1: What is interactive programming?")
print(GREEN + "A1: Running a program directly in the terminal to see instant output." + RESET)
print(GREEN + "Example: typing 'print(5+3)' in Python shell\n" + RESET)
```

# Question 2

```python
print("Q2: What is script programming?")
print(GREEN + "A2: Saving code in a file and running it with 'python filename.py'." + RESET)
print(GREEN + "This is useful for bigger programs.\n" + RESET)
```

# Question 3

```python
print("Q3: What is a variable?")
print(GREEN + "A3: A variable stores data in memory with a name." + RESET)
print(GREEN + "Example: x = 10, name = 'Vaanya'\n" + RESET)
```

# Question 4

```python
print("Q4: What is a string?")
print(GREEN + "A4: A string is text in single or double quotes." + RESET)
print(GREEN + "Example: 'Hello' or \"World\"\n" + RESET)
```

# Question 5

```python
print("Q5: What are numbers?")
print(GREEN + "A5: Numbers can be integers (int) or decimals (float)." + RESET)
print(GREEN + "Example: a = 5, b = 3.2\n" + RESET)
```

# Question 6

```python
print("Q6: What is a list?")
print(GREEN + "A6: A list stores multiple items in order, can be changed (mutable)." + RESET)
print(GREEN + "Example: fruits = ['apple', 'banana']\n" + RESET)
```

# Question 7

```python
print("Q7: What is a tuple?")
print(GREEN + "A7: A tuple is like a list, but cannot be changed (immutable)." + RESET)
print(GREEN + "Example: point = (2, 3)\n" + RESET)
```
